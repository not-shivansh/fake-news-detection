```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re 
import string
```

```python
d_fake = pd.read_csv('Fake.csv')
d_true = pd.read_csv('True.csv')
```

```python
d_fake.head()
```

```python
d_true.head()
```

```python
d_fake["class"] = 0
d_true["class"] = 1
```

```python
d_fake.shape, d_true.shape
```

```python
d_fake_man_test = d_fake.tail(10)
for i in range(23480,23470,-1):
    d_fake.drop([i],axis =0, inplace = True)

d_true_man_test = d_true.tail(10)
for i in range(21416,21406,-1):
    d_true.drop([i],axis =0, inplace = True ) 
```

```python
d_fake.shape,d_true.shape
```

```python
d_fake_man_test["class"]=0
d_true_man_test["class"]=1
```

```python
d_fake_man_test.head
```

```python
d_true_man_test.head
```

```python
data_merge = pd.concat([d_fake,d_true],axis=0)
data_merge.head
```

```python
data_merge.columns
```

```python
data = data_merge.drop(['title','subject','date'],axis=1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac=1)
```

```python
data.head
```

```python
data.reset_index(inplace=True)
data.drop(['index'],axis=1,inplace=True)
```

```python
data.columns
```

```python
data.head
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]','',text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+','',text)
    text = re.sub('<.*?>+','',text)
    text = re.sub('[%s]'%re.escape(string.punctuation),'',text)
    text = re.sub('\n','',text)
    text = re.sub('\w*\d\w*','',text)
    return text
```

```python
data['text']=data['text'].apply(wordopt)
```

```python
x = data['text']
y = data['class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
DT.fit(xv_train,y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GB = GradientBoostingClassifier(random_state=0)
GB.fit(xv_train,y_train)
```

```python
predict_gb = GB.predict(xv_test)
```

```python
GB.score(xv_test,y_test)
```

```python
print(classification_report(y_test,predict_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(random_state=0)
RF.fit(xv_train,y_train)
```

```python
predict_rf = RF.predict(xv_test)
```

```python
RF.score(xv_test,y_test)
```

```python
print(classification_report(y_test,predict_rf))
```

```python
def output_label(n):
    if n==0:
        return "Fake News"
    elif n==1:
        return "Not A Fake News"
    
def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] =  new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_lr = LR.predict(new_xv_test)
    pred_dt = DT.predict(new_xv_test)
    predict_gb = GB.predict(new_xv_test)
    predict_rf = RF.predict(new_xv_test)
    return print("\n\nLR Prediction:{}\nDT Prediction:{}\nGB Prediction:{}\nRF Prediction:{}".format(output_label(pred_lr[0]), output_label(pred_dt[0]), output_label(predict_gb[0]), output_label(predict_rf[0])))
```
